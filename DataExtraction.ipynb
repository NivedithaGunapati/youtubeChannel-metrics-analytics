```python
import pandas as pd
import gspread
from oauth2client.service_account import ServiceAccountCredentials
from gspread_dataframe import set_with_dataframe

# Define the scope
scope = ['https://spreadsheets.google.com/feeds',
         'https://www.googleapis.com/auth/drive']

def export_to_google_sheets(df, sheet_url, credentials_file):
    # Load Google Sheets credentials from JSON file
    credentials = ServiceAccountCredentials.from_json_keyfile_name(credentials_file, scope)

    # Authorize the client
    client = gspread.authorize(credentials)

    # Open the Google Spreadsheet by URL
    spreadsheet = client.open_by_url(sheet_url)

    # Select the first sheet
    worksheet = spreadsheet.sheet1

    # Clear existing content (optional)
    worksheet.clear()

    # Export DataFrame to Google Sheets
    set_with_dataframe(worksheet, df)

    # Open the Google Spreadsheet in browser
    import webbrowser
    webbrowser.open(sheet_url)

# Example usage:
df = pd.DataFrame({
    'Title': titles,
    'Views': views,
    'Likes': likes,
    'Published': published
})

# Specify the URL of the Google Spreadsheet
sheet_url = 'https://docs.google.com/spreadsheets/d/1MUee9QI1q0VuJtDoGhca_H2C1AqV0fHNsxRjZdYh3ao/edit#gid=0'

# Path to the service account credentials file
credentials_file = 'data-414904-a81be48441c1.json'

export_to_google_sheets(df, sheet_url, credentials_file)
```

**How to create API credentials**

Step 1: Set Up Google Cloud Platform (GCP) Project
Go to the Google Cloud Console.

Create a new project or select an existing project where you want to enable the Google Sheets API.

Step 2: Enable the Google Sheets API
In the Cloud Console, go to the "APIs & Services" -> "Library" section.

Search for "Google Sheets API" and enable it for your project.

Step 3: Create Service Account and Generate Credentials

In the Cloud Console, go to the "APIs & Services" -> "Credentials" section.

Click on "Create credentials" and select "Service account".

Fill in the necessary details for the service account (e.g., name, description) and click on "Create".

Assign the "Project -> Editor" role to the service account.

Click on "Continue" and then "Done".

After creating the service account, click on the service account email and go to the "Keys" tab.

Click on "Add Key" -> "Create new key" and select "JSON" format to download the credentials file (service_account_key.json).

Step 4: Find the Email Address of the Service Account:

Locate the email address of the service account you created in the Google Cloud Console. It usually follows the pattern service-account@your-project-id.iam.gserviceaccount.com.

Open the Google Spreadsheet:

Go to Google Drive (drive.google.com) and open the Google Spreadsheet you want to share with the service account.

Click on the "Share" Button:

In the top right corner of the Google Spreadsheet, click on the "Share" button.
Add the Email Address of the Service Account:

In the "Share with people and groups" dialog box, paste the email address of the service account into the "Add people and groups" field.
Set Permissions:

Select the desired permissions for the service account. For example, you can give it "edit" access to allow it to modify the spreadsheet data.
Send the Invitation:

Click on the "Send" button to send the invitation to the service account.
